```python
import copy

import numpy as np
import torch

from core.clustering import Hierarchical_Clustering
from core.clustering import SHAPE

seed = 370
torch.manual_seed(seed)
np.random.seed(seed)
n_clients = 10
model_features = 10000
weights_list = torch.randn(n_clients, model_features)
samples = np.random.randint(100, 10000, size=(n_clients, 1)).squeeze().tolist()
model = Hierarchical_Clustering(weights_list, samples, threshold=0.03, increment_factor=1.1, device="cpu")
model.fit()
print("Root:", model.root())
print(model)
model.visualize_tree(save_plot=False, show=True)
```

```python
weights_list = torch.randn(n_clients, model_features)
model.update_clients_centroids(weights_list)
print(model)
model_1 = copy.deepcopy(model)
model_2 = copy.deepcopy(model)

shape_1 = SHAPE(tree=model_1)
shape_2 = SHAPE(tree=model_2)
```

```python
a = model_1.nodes[10]
b = model_1.nodes[11]
shape_1.merge([a, b])
model_1.visualize_tree(save_plot=False)
```

```python
import random

def generate_mock_data(num_rounds=30, total_client_pool=15):
    """
    生成模拟联邦学习验证准确率的数据。
    结构: List[Dict[client_id, accuracy]]
    """
    val_accuracies = []

    for r in range(num_rounds):
        round_data = {}

        # 1. 模拟模型性能随时间提升的基准线 (从 0.2 线性提升到 0.85)
        base_accuracy = 0.2 + (0.65 * (r / num_rounds))

        # 遍历潜在的客户端池
        for client_id in range(total_client_pool):

            # --- 模拟特殊情况 ---

            # 情况 A: 节点 2 在第 15 轮后永久掉线
            if client_id == 2 and r > 15:
                continue

            # 情况 B: 节点 5 是新加入的节点，第 10 轮才开始出现
            if client_id == 5 and r < 10:
                continue

            # --- 模拟随机参与 (Partial Participation) ---
            # 假设每轮每个节点有 60% 的概率被选中参与训练/验证
            # 节点 0 设为 Always Online 以作为参考
            if client_id == 0 or random.random() > 0.4:

                # --- 模拟准确率波动 ---
                # 每个 Client 有一点自己的偏差 (有的数据好，有的数据差)
                client_bias = (client_id % 5 - 2) * 0.03  # -0.06 到 +0.06

                # 随机噪声
                noise = random.uniform(-0.02, 0.02)

                # 计算最终准确率
                acc = base_accuracy + client_bias + noise

                # 限制在 0.0 - 1.0 之间
                acc = max(0.0, min(1.0, acc))

                round_data[client_id] = acc

        val_accuracies.append(round_data)

    return val_accuracies

# --- 使用示例 ---

# 1. 生成数据
mock_stats_data = generate_mock_data()

# 2. 打印查看结构 (打印前3轮)
print("数据结构示例 (前3轮):")
for i, r_data in enumerate(mock_stats_data[:3]):
    print(f"Round {i}: {r_data}")

# 3. 如果你要在你的类中测试，可以这样赋值:
# self.stats["val_accuracies"] = mock_stats_data
# self.visualize_val_accuracy(save_plot=False)

```

```python
import random

def generate_mock_train_losses(num_rounds=5, total_clients=5):
    """
    生成结构: List[Dict[client_id, List[float]]]
    每个Client每轮会有 3-10 个 local updates (loss values)
    """
    train_losses = []

    for r in range(num_rounds):
        round_data = {}

        # 模拟 Loss 随轮次下降的趋势
        base_loss = 2.5 * (0.9 ** r)

        for client_id in range(total_clients):
            # 模拟部分参与：每轮只有 70% 概率参与
            if random.random() > 0.3:

                # 每个 Client 本地更新次数不一定相同 (例如 3 到 8 次 batch update)
                num_local_updates = random.randint(3, 8)

                client_losses = []
                current_loss = base_loss + random.uniform(-0.2, 0.2)

                # 模拟在一个 Round 内，Loss 也在逐渐下降
                for step in range(num_local_updates):
                    current_loss = current_loss * 0.95 + random.uniform(-0.05, 0.05)
                    client_losses.append(max(0.0, current_loss))

                round_data[client_id] = client_losses

        train_losses.append(round_data)

    return train_losses

# --- 使用测试 ---
# self.stats["train_losses"] = generate_mock_train_losses()
# self.visualize_train_losses(save_plot=False)

# --- 使用示例 ---

# 1. 生成数据
mock_stats_data = generate_mock_train_losses()

# 2. 打印查看结构 (打印前3轮)
print("数据结构示例 (前3轮):")
for i, r_data in enumerate(mock_stats_data[:3]):
    print(f"Round {i}: {r_data}")

# 3. 如果你要在你的类中测试，可以这样赋值:
# self.stats["val_accuracies"] = mock_stats_data
# self.visualize_val_accuracy(save_plot=False)
```
